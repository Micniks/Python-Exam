```python
data = pd.read_csv('https://raw.githubusercontent.com/oganm/dndstats/master/docs/charTable.tsv', sep='\t')

#Get List of single class strings, not including any multiclasses
classes_list = data[['justClass']].drop_duplicates()
single_class_list = [c for c in classes_list['justClass'] if '|' not in c]
single_class_list.sort()

#Get List of single race strings, removing subtypes
full_race_data = data[['race']].drop_duplicates()
full_race_list = list(dict.fromkeys([r.split()[len(r.split())-1] for r in full_race_data['race'] if isinstance(r,str)]))
full_race_list.sort()

#display(single_class_list)
display(full_race_list)
```
